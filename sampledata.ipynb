```r
data <- read.csv("D:\\projectR\\datafiles\\creditcard.csv", header = TRUE,
                 stringsAsFactors = FALSE,
                 strip.white = TRUE ,
                 sep = ',')
```

```r
library("sqldf")
data1 <- sqldf("select * from data where Class=1")
data0 <- sqldf("select * from data where Class=0")
nrow(data1)
```

```r
library("dplyr")
set.seed(400)
data0 <- sample_n(data0, 492*3)
nrow(data0)
```

```r
data.final <- rbind(data0,data1)
nrow(data.final)
set.seed(400)
data.final <- data.final[sample(nrow(data.final)),]
head(data.final)
```

```r
write.csv(data.final,file="D:\\projectR\\datafiles\\sampledata.csv")
```
