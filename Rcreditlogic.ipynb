```r
data <- read.csv("C:\\Users\\akhil\\Desktop\\creditcard.csv", header = TRUE,
                 stringsAsFactors = FALSE,
                 strip.white = TRUE ,
                 sep = ',')







```

```r
dm <-scale(data)
dm <- data[c(2,5,9,10,11,14,15,21,22,23,28,29,30,31)]
head(dm)
summary(dm)
```

```r
library("dplyr")
dm <- dm %>% 
  mutate(combcol = V1+V4+V8+V9+V10+V13+V14+V20+V21+V22+V27+V28)
head(dm)

fit = glm(Class ~ combcol, data=dm, family=binomial)
summary(fit)
```

```r
newdat <- data.frame(
  combcol =seq(min(dm$combcol), max(dm$combcol),
               
               len=100))

#newdat

newdat$Class = predict(fit, newdata=newdat, type="response")

plot(Class ~ combcol, data=dm, col="red4")

lines(Class ~ combcol, newdat, col="green4", lwd=2)
```

```r
library(InformationValue)
predicted <- predict(fit, dm, type="response")  # predicted scores
#predicted
plotROC(data, predicted)
```

```r
sensitivity(dm$Class, predicted, threshold = 0.8)

specificity(dm$Class, predicted, threshold = 0.8)


confusionMatrix(dm$Class, predicted > 0.0015)
```
