```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("D:\\newproject\\New folder\\Barley.data.csv")
```

```python
#Na Handling
df.isnull().values.any()
```

```python
from sklearn.model_selection import cross_val_predict, cross_val_score
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
X = df.drop(['Predictor'], axis=1)
X_col = X.columns
```

```python
y = df['Predictor']
```

```python

#Savitzky-Golay filter with second degree derivative.
from scipy.signal import savgol_filter 

sg=savgol_filter(X,window_length=11, polyorder=3, deriv=2, delta=1.0)
```

```python
sg_x=pd.DataFrame(sg, columns=X_col)
sg_x.head()
```

```python
from sklearn.decomposition import PCA as sklearnPCA
sklearn_pca = sklearnPCA(n_components=100)
X_sklearn = sklearn_pca.fit_transform(sg_x)
```

```python
df15 = pd.DataFrame(X_sklearn)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_sklearn, y,
                                                    train_size=0.7,
                                                    random_state=42,stratify = y)
```

```python
def print_score(clf, X_train, y_train, X_test, y_test, train=True):
    if train:
        print("Train Result:\n")
        print("accuracy score: {0:.4f}\n".format(accuracy_score(y_train, clf.predict(X_train))))
        print("Classification Report: \n {}\n".format(classification_report(y_train, clf.predict(X_train))))
        print("Confusion Matrix: \n {}\n".format(confusion_matrix(y_train, clf.predict(X_train))))

        res = cross_val_score(clf, X_train, y_train, cv=10, scoring='accuracy')
        print("Average Accuracy: \t {0:.4f}".format(np.mean(res)))
        print("Accuracy SD: \t\t {0:.4f}".format(np.std(res)))
        
    elif train==False:
        print("Test Result:\n")        
        print("accuracy score: {0:.4f}\n".format(accuracy_score(y_test, clf.predict(X_test))))
        print("Classification Report: \n {}\n".format(classification_report(y_test, clf.predict(X_test))))
        print("Confusion Matrix: \n {}\n".format(confusion_matrix(y_test, clf.predict(X_test)))) 
```

```python
from sklearn import svm
clf = svm.SVC(kernel="linear")
clf.fit(X_train,y_train)
```

```python
print_score(clf, X_train, y_train, X_test, y_test, train=True)
```

```python
print_score(clf, X_train, y_train, X_test, y_test, train=False)
```

```python
from sklearn.ensemble import RandomForestClassifier
clf1 = RandomForestClassifier(random_state=101)
clf1.fit(X_train,y_train)
```

```python
print_score(clf1, X_train, y_train, X_test, y_test, train=True)
```

```python
print_score(clf1, X_train, y_train, X_test, y_test, train=False)
```

```python
# Importing decision tree classifier from sklearn library
from sklearn.tree import DecisionTreeClassifier

# Fitting the decision tree with default hyperparameters, apart from
# max_depth which is 5 so that we can plot and read the tree.
dt_default = DecisionTreeClassifier()
dt_default.fit(X_train, y_train)
```

```python
from sklearn.ensemble import RandomForestClassifier
clf1 = RandomForestClassifier(random_state=42)

# Importing decision tree classifier from sklearn library
from sklearn.tree import DecisionTreeClassifier

# Fitting the decision tree with default hyperparameters, apart from
# max_depth which is 5 so that we can plot and read the tree.
dt_default = DecisionTreeClassifier()

from sklearn import svm
clf = svm.SVC(kernel="linear")
clf.fit(X_train,y_train)

from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=4,
                           p=2, metric='minkowski')

for i in range(1, 100, 1):
    from sklearn.decomposition import PCA as sklearnPCA
    sklearn_pca = sklearnPCA(n_components=i)
    X_sklearn = sklearn_pca.fit_transform(sg_x)
    df15 = pd.DataFrame(X_sklearn)
    X_train, X_test, y_train, y_test = train_test_split(X_sklearn, y,train_size=0.8,random_state=42,stratify = y)
    clf1 = clf1.fit(X_train,y_train)
    #print("i = {0}".format(i))
    #print("random forest")
    #print("accuracy score train: {0:.4f}\n".format(accuracy_score(y_train, clf1.predict(X_train))))
    if accuracy_score(y_test, clf1.predict(X_test)) >0.8:
        print("random_forest i = {0}".format(i))
        print("accuracy score test: {0:.4f}\n".format(accuracy_score(y_test, clf1.predict(X_test))))
    des = dt_default.fit(X_train, y_train)
    #print("decision tree")
    #print("accuracy score train: {0:.4f}\n".format(accuracy_score(y_train, des.predict(X_train))))
    if accuracy_score(y_test, des.predict(X_test)) >0.8:
        print("decision_tree i = {0}".format(i))
        print("accuracy score test: {0:.4f}\n".format(accuracy_score(y_test, des.predict(X_test))))
    clf=clf.fit(X_train,y_train)
    #print("svm")
    #print("accuracy score train: {0:.4f}\n".format(accuracy_score(y_train, clf.predict(X_train))))
    if accuracy_score(y_test, clf.predict(X_test)) >0.8:
        print("svm i = {0}".format(i))
        print("accuracy score test: {0:.4f}\n".format(accuracy_score(y_test, clf.predict(X_test))))
    knn =knn.fit(X_train, y_train)
    #print("knn")
    #print("accuracy score train: {0:.4f}\n".format(accuracy_score(y_train, knn.predict(X_train))))
    if accuracy_score(y_test, knn.predict(X_test)) >0.8:
        print("knn i = {0}".format(i))
        print("accuracy score test: {0:.4f}\n".format(accuracy_score(y_test, knn.predict(X_test))))
    
```
